```python
import json
import pandas as pd
```

```python
with open('04_iot_data/device_readings.json') as js:
    data = json.load(js)

rows = []
for rec in data:
    timestamp = rec["timestamp"]
    for reading in rec["readings"]:
        rows.append({
            "timestamp": timestamp,
            "sensor_id": reading["sensor_id"],
            "distance_m": reading["distance_m"]
        })

data_csv = pd.DataFrame(rows)
data_csv.to_csv('device_readings.csv', index=False)
```
